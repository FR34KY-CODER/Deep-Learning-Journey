#Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

#Framing a Toy Dataset

```python
x,y = make_classification(n_samples=100, n_features=2,n_informative=1,n_redundant=0,n_classes= 2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
#only class_sep is mostly valuable as it tells the data of features to be more or less distinguishable
```

####Plot representation of dataset

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y, cmap= 'winter',s =100)
plt.show()
```

#Defining a Perceptron

```python
def perceptron(x,y):

  x = np.insert(x,0,1,axis=1) #axis =1 is the bias 'b'
  weights= np.ones(x.shape[1])
  lr = 0.1 #learning rate
  for i in range (200): #epoch value
    j = np.random.randint(0,100) #randomly choosing a node/element
    y_hat = step(np.dot(x[j],weights))
    weights = weights + lr*(y[j] - y_hat)*x[j] #pereceptron trick algorithm
  return weights[0],weights[1:]
```

```python
def step(z):
  return 1 if z>0 else 0
```

#Defining the Straight Line Constants (Running the Algorithm)

```python
intercept_,coef_ = perceptron (x,y)
```

```python
m = -(coef_[0]/coef_[1])    #straight line slope coefficient of x
b = -(intercept_/coef_[1])  #b = c which is y intercept of standard straight line equation
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*(x_input) + b
```

#Distributing Line of Differnece created by Perceptron

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y, cmap= 'winter',s =100)
plt.plot(x_input,y_input, color='red', linewidth=3)
plt.ylim(-3,2)
```
